# Linear Regression with Honey Production

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

df = pd.read_csv("honeyproduction.csv")
```

```python
print(df.head())
```

```python
prod_per_year = df.groupby('year').totalprod.mean().reset_index()
print(prod_per_year.head())
```

```python
X = prod_per_year['year']
X = X.values.reshape(-1,1) 
Y = prod_per_year['totalprod']

plt.scatter(X, Y)
plt.xlabel('Years')
plt.ylabel('Total Production (average)')
plt.title('Honey Production over time')
plt.show()
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, Y)
print(regr.coef_[0])
```

```python
y_predict = regr.predict(X)
plt.plot(X, y_predict)
plt.xlabel('Years')
plt.ylabel('Total Production (average)')
plt.title('Honey Production over time')
plt.show()

```

Predicting Honey Production in 2050 

```python
X_future = np.array(range(2013, 2050))
X_future = X_future.reshape(-1,1)
future_predict = regr.predict(X_future)
plt.xlabel('Years')
plt.ylabel('Total Production (average)')
plt.title('Honey Production by 2050')
plt.plot(X_future, future_predict)
plt.show()
```
